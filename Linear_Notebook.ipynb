```python
#So, a linear regression its basically a linear model that represents the relationship between a scalar response and one
# or more explanatory variables
#the whole point of this notebook its only to give him a test, not to make a complex 

#Fuck useful shit, all my homies use unaccurate linear regression
from sklearn.linear_model import LinearRegression
import random
```

```python
#our '''''data sets'''
feature_set = []
target_set = []

#setting our number of rows
number_of_rows = 200
#limitting the possible values in our data set
random_number_limit = 2000
```

```python
#our data set
for i in range(0, number_of_rows):
    x = random.randint(0, random_number_limit)
    y = random.randint(0, random_number_limit)
    z = random.randint(0, random_number_limit)
    
    #now, we are going to create our  linear fuction for the target data set
    function = (10 * x)+(2 * y) + (3*z)
    #data append
    feature_set.append([x,y,z])
    target_set.append(function)
```

```python
target_set
```

```python
#now that we have our data sets its time to cretate our model
model = LinearRegression()
model.fit(feature_set, target_set)
```

```python
#time to test
test_set = [[9,5,0]] # we should see in our output = function(9,6,0)= (10 * 9) + (2 * 5) + (3 * 0) = 100
prediction = model.predict(test_set)
```

```python
print('Prediciton: ' + str(prediction)+' Coefficients: ' + str(model.coef_))
```
